```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from IPython.core import display as ICD
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
file_path = 'C:/Users/James Sloan/Documents/School/IE 4910 Python ML/Assignments/PCC5/PCC5 - Datasets/'
train = pd.read_csv(file_path + 'Spain Rail Transportation.csv')
```

```python
#Convert date time columns to date time object
cols = ['insert_date','start_date','end_date']
for col in cols:
    train[col] = pd.to_datetime(train[col], infer_datetime_format=True)

#Add trip length column
train['trip_length'] = (train['end_date']-train['start_date'])

#Drop useless first column and insert date column
train = train[['origin', 'destination', 'start_date', 
               'end_date', 'trip_length', 'train_type',
               'price', 'train_class', 'fare']]

#Check data types and format
print(train.dtypes)
print(train.shape)
train.head(1)
```

```python
#check for Nan values
null = pd.DataFrame(train.isnull().sum())
print('Count of Null Data Points:')
display(null)
```

```python
#Drop all Nans. Mostly blanks, really only losing ~3k results
train.dropna(axis=0, how='any', inplace=True)
print(train.shape)

#recheck for Nan values
null = pd.DataFrame(train.isnull().sum())
print('Count of Null Data Points:')
display(null)
```

```python
#Stat summary
train.describe(include = 'all')
```

```python
#Plot average trip length
import matplotlib.pyplot as plt

#set index to start date
trainplot = train[['start_date','trip_length']]
trainplot = trainplot.set_index('start_date')

#Create avg time column
trainplot['avg_delta'] = np.mean(train['trip_length'])
trainplot['avg_delta'] = trainplot['avg_delta']/3600000000000
trainplot['trip_length'] = trainplot['trip_length']/3600000000000

#Overall plot
trainplot.plot(figsize = (10,10))
plt.xlabel('Start Date')
plt.ylabel('Trip Length')
plt.title('Length of train rides over time')
```

```python
#Create month, week, and day columns
train['date'] = train['start_date']
train = train.set_index('date')
train['month'] = train.index.month
train['week'] = train.index.week
train['day'] = train.index.day
train['hour'] = train.index.hour
#train.head(5)
```

```python
from sklearn.preprocessing import scale

#drop repetitive columns
train = train.drop(['trip_hours','start_date','end_date'], axis=1)
train['trip_length'] = train['trip_length']/pd.Timedelta('1 hour')

#set x and y
x = train.drop(['fare'], axis=1)
y = train['fare']

#Binary encode categorical data
cols = ['origin','destination','train_type','train_class']
x = pd.get_dummies(x, columns = cols)

#scale x
x = scale(x)
```

```python
#test c values at varied test ratio
from sklearn import model_selection
from sklearn.svm import SVC

SVC_score = {}
for test_ratio in range(10,100,10):
    for c in range(0,101,10):
        X_train,X_test,y_train,y_test = model_selection.train_test_split(x, y,
                                                        test_size=test_ratio/100, 
                                                        random_state=42)
        if c == 0:
            c = 1
        svm_model = SVC(kernel='poly', degree=2, coef0=1, C=c)
        svm_model.fit(X_train,y_train)
        name = ('ratio:{}%, C:{}'.format(str(test_ratio),str(c)))
        SVC_score[name] = svm_model.score(X_test,y_test)

for key in SVC_score.keys():
    print('({}) score: {}'.format(key,SVC_score[key]))
```

```python
sen_x = []
sen_y = []

c = 50

for test_ratio in range(10,100,10):
    X_train,X_test,y_train,y_test = model_selection.train_test_split(x, y,
                                                test_size=test_ratio/100, 
                                                random_state=42)
    svm_model = SVC(kernel='poly', degree=2, coef0=1, C=c)
    svm_model.fit(X_train,y_train)
    sen_x.append(test_ratio)
    sen_y.append(svm_model.score(X_test,y_test))

plt.plot(sen_x,sen_y)
plt.xlabel('Test %')
plt.ylabel('Model Score')
plt.title('Sensitivity Analysis (C = 50)')
```

```python
# Set model at 60% test and c = 50

X_train,X_test,y_train,y_test = model_selection.train_test_split(x, y,
                                            test_size=0.6, 
                                            random_state=42)
svm_model = SVC(kernel = 'poly', degree = 2, coef0 = 1, C = 50)
svm_model.fit(X_train,y_train)
```

```python
#Decision tree parameters
from sklearn.tree import DecisionTreeClassifier

depth_score = {}
for depth in range(1,11):
    dt_model = DecisionTreeClassifier(max_depth = depth, 
                                      random_state = 42)
    dt_model.fit(X_train,y_train)
    depth_score[depth] = dt_model.score(X_test,y_test)
    
depths = []
dscores = []
for key in depth_score.keys():
    depths.append(key)
    dscores.append(depth_score[key])

plt.plot(depths,dscores)
plt.xlabel('Max Depth')
plt.ylabel('Model Score')
plt.title('Max depth parameter analysis (test size: 60%)')
```

```python
# Set model to max_depth = 8
dt_model = DecisionTreeClassifier(max_depth = 8, 
                                  random_state = 42)
dt_model.fit(X_train,y_train)
depth_score[depth] = dt_model.score(X_test,y_test)
```

```python
#Random forest parameters
#WARNING: THIS TAKES QUITE A WHILE TO RUN
from sklearn.ensemble import RandomForestClassifier

n_ = []
mn_ = []
score_ = []

for n in range (0,1000,100):
    if n == 0:
        n=1
    for max_l_n in range (5,55,5):
        n_.append(n)
        mn_.append(max_l_n)
        rf_model = RandomForestClassifier(n_estimators = n, 
                                         max_leaf_nodes = max_l_n,
                                         n_jobs = 1,
                                         random_state = 42)
        rf_model.fit(X_train, y_train)
        score_.append(rf_model.score(X_test,y_test))
```

```python
#plot RF parameters
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
ax.plot_trisurf(n_, mn_, score_, cmap='RdYlGn')
plt.xlabel('num estimators')
plt.ylabel('max leaf nodes')
plt.title('RF model score')
```

```python
#Set rf model to num est = 200, max leaf nodes = 20

rf_model = RandomForestClassifier(n_estimators = 200, 
                                max_leaf_nodes = 20,
                                n_jobs = 1,
                                random_state = 42)
rf_model.fit(X_train, y_train)
```

```python
#Report all for each model type with best params
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

models = [svm_model, dt_model, rf_model]
name = {svm_model:'SVM',
        dt_model:'DT',
        rf_model:'RF'}
confusion = {}
report = {}
model_score = {}
for model in models:
    prediction = model.predict(X_test)
    confusion[name[model]] = confusion_matrix(y_test, prediction)
    report[name[model]] = classification_report(y_test,prediction)
    model_score[name[model]] = model.score(X_test, y_test)
    
for model in name.values():
    print('{} model:'.format(model))
    print('Confusion matrix:\n{}'.format(confusion[model]))
    print('Classification report:\n{}\n\n\n'.format(report[model]))
```
