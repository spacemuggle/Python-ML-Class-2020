```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from IPython.core import display as ICD
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
#read in data, convert date to datetime object, set index, make copy to play with
AMZNoriginal = pd.read_csv('AMZN.csv')
AMZNoriginal['Date'] = pd.to_datetime(AMZNoriginal['Date'], format='%m/%d/%y')
AMZNoriginal = AMZNoriginal.set_index('Date')
AMZN = AMZNoriginal.copy(deep=True)


display(AMZN.head(10))

# print('(# Row, # Col)',AMZN.shape)
```

```python
#Check data types and null values to prevent surprises
print(AMZN.dtypes)
null = pd.DataFrame(AMZN.isnull().sum())
print('\nNull Data Points:')
display(null)
```

```python
#Use histograms to see the general spread of the data
AMZN.hist(bins=75, figsize=(10,10))
plt.show()
```

```python
#Extrapolate date data for later analysis
AMZN['Year'] = AMZN.index.year
AMZN['Month'] = AMZN.index.month
AMZN['Day'] = AMZN.index.day
AMZN['WeekdayName'] = AMZN.index.weekday_name

# AMZN.loc['1999-01-05':'1999-01-20'] This is super cool, can even do by year or month
```

```python
#Look at the range of data
StartDate = AMZN.index[0]
print(StartDate)

EndDate = AMZN.index[-1]
print(EndDate)
```

```python
#Create monthly and yearly average data
MonthlyStats = AMZN.groupby(['Year','Month'])['Adj Close'].mean()
YearlyStats = AMZN.groupby(['Year'])['Adj Close'].mean()
#Create monthly and yearly deltas
MonthlyDelta = MonthlyStats.diff()
YearlyDelta = YearlyStats.diff()

#display(MonthlyStats.head(10))
#display(YearlyStats.head(10))
#display(MontlyDelta.head(10))
#display(YearlyDelta.head(10))

AMZN['Date'] = AMZN.index
#AMZN.head(10)
```

```python
#Merge Monthly and Yearly averages back into main df

cols = ['Year', 'Month'] 
AMZN = AMZN.merge(MonthlyStats, left_on=cols, right_on=cols, how='left')
cols = ['Year']
AMZN = AMZN.merge(YearlyStats, left_on=cols, right_on=cols, how='left')

AMZN.rename(columns={'Adj Close': 'YearAvg',
                     'Adj Close_y': 'MonthAvg',
                     'Adj Close_x': 'AdjClose'}, inplace=True)
```

```python
#Merge Monthly and Yearly Deltas back into main df
cols = ['Year', 'Month'] 
AMZN = AMZN.merge(MonthlyDelta, left_on=cols, right_on=cols, how='left')
cols = ['Year']
AMZN = AMZN.merge(YearlyDelta, left_on=cols, right_on=cols, how='left')

AMZN.rename(columns={'Adj Close_x': 'DeltaMonth',
                     'Adj Close_y': 'DeltaYear'}, inplace=True)
print(AMZN.columns)
# print(AMZN.shape)
```

```python
#Resets index to date after merge
AMZN = AMZN.set_index('Date') #Move down if more merges needed
```

```python
#Create delta day
AMZN['DeltaDay'] = AMZN['AdjClose'].diff()

#Create rolling average data and normalize adj close to those values
AMZN['7DayMA'] = AMZN.rolling(7).mean()['AdjClose']
AMZN['30DayMA'] = AMZN.rolling(30).mean()['AdjClose']
AMZN['90DayMA'] = AMZN.rolling(90).mean()['AdjClose']
AMZN['NormedOn7MA'] = AMZN['AdjClose'] / AMZN['7DayMA']
AMZN['NormedOn30MA'] = AMZN['AdjClose'] / AMZN['30DayMA']
AMZN['NormedOn90MA'] = AMZN['AdjClose'] / AMZN['90DayMA']

display(AMZN.head(10))
```

```python
#1. Create a time series of the data provided
#entire range
plotAMZN = AMZN[['AdjClose','MonthAvg','YearAvg','7DayMA','30DayMA','90DayMA']]
plotAMZN.plot(figsize=(10,6),title='1997-2019')
#1997-1999
plotAMZN.loc['1997':'1999'].plot(figsize=(10,6),title='1997-1999')
#2000-2004
plotAMZN.loc['2000':'2004'].plot(figsize=(10,6),title='2000-2004')
#2005-2009
plotAMZN.loc['2005':'2009'].plot(figsize=(10,6),title='2005-2009')
#2010-2014
plotAMZN.loc['2010':'2014'].plot(figsize=(10,6),title='2010-2014')
#2015-2019
plotAMZN.loc['2015':'2019'].plot(figsize=(10,6),title='2015-2019')
```

```python
#2. Create a statistical summary of the data set
display(AMZN.describe())
```

```python
#looking for correlation with close
corr_matrix = AMZN.corr()
display(corr_matrix['AdjClose'].sort_values(ascending=False))

#looking for correlation with delta day
display(corr_matrix['DeltaDay'].sort_values(ascending=False))

#looking for correlation with delta month
display(corr_matrix['DeltaMonth'].sort_values(ascending=False))

#looking for correlation with delta year
display(corr_matrix['DeltaYear'].sort_values(ascending=False))
```

```python
from sklearn.model_selection import train_test_split

#fill na in delta year
AMZN.loc['1997','DeltaYear']=0
#Cut out first 90 days to be able to use MAs
AMZNlearn = AMZN.loc['1997-9-22':].reset_index()
#Get rid of recurring date fields
AMZNlearn = AMZNlearn[['Open','High','Low','Close','AdjClose',
                       '7DayMA','30DayMA','90DayMA',
                       'YearAvg','MonthAvg',
                       'DeltaYear','DeltaMonth','DeltaDay']]
```

```python
#display(AMZNlearn.tail(10))

#only using non-forward looking data to try and create useful algorithm
coefs1 = ['7DayMA','30DayMA','90DayMA','DeltaYear','DeltaMonth','DeltaDay']
x = AMZNlearn[coefs1]
objective1 = 'AdjClose'
y = AMZNlearn[objective1] 

#second set based on predicting yearly price change
coefs2 = ['7DayMA','30DayMA','90DayMA','DeltaMonth','DeltaDay']
x2 = AMZNlearn[coefs2]
objective2 = 'DeltaYear'
y2 = AMZNlearn[objective2]
```

```python
#Split model into training and testing set
from sklearn import model_selection

X_train,X_test,y_train,y_test = model_selection.train_test_split(x, y,
                                                                 test_size=0.2, random_state=42)

X2_train,X2_test,y2_train,y2_test = model_selection.train_test_split(x2, y2,
                                                                 test_size=0.2, random_state=42)

```

```python
#Standardize features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)

X2_train_std = scaler.fit_transform(X2_train)
X2_test_std = scaler.fit_transform(X2_test)
```

```python
display(X_train_std)
print(X_train_std.shape)
display(X2_train_std)
print(X2_train_std.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import Lasso

#Create basic linear reg as baseline
lin_reg = LinearRegression()
lin_reg.fit(X_train_std, y_train)

#Create lasso regression
las_reg = Lasso()
las_reg.fit(X_train_std, y_train)

#Create Ridge regression with 3 different alphas
rig_reg_CV = RidgeCV(alphas = [0.1, 1, 10])
model_CV = rig_reg_CV.fit(X_train_std, y_train)
```

```python
#view best alpha value
print(model_CV.alpha_)
```

```python
from sklearn.linear_model import Ridge

#set Ridge reg to use best alpha value
rig_reg = Ridge(alpha = 0.1)
rig_reg_model = rig_reg.fit(X_train_std, y_train)
#rig_reg_model2 = rig_reg.fit(X2_train_std, y2_train)
```

```python
#Calculate training and testing scores
Scores = {}
Scores['Linear Train Score'] = lin_reg.score(X_train_std, y_train)
Scores['Linear Test Score'] = lin_reg.score(X_test_std, y_test)

Scores['Ridge Train Score'] = rig_reg.score(X_train_std, y_train)
Scores['Ridge Test Score'] = rig_reg.score(X_test_std, y_test)

Scores['Lasso Train Score'] = las_reg.score(X_train_std, y_train)
Scores['Lasso Test Score'] = las_reg.score(X_test_std, y_test)
print(Scores)
```

```python
#display training vs. testing scores
print('Linear regression training score: {}'.format(Scores['Linear Train Score']))
print('Linear regression testing score : {}'.format(Scores['Linear Test Score']))
print('\nRidge regression training score : {}'.format(Scores['Ridge Train Score']))
print('Ridge regression testing score  : {}'.format(Scores['Ridge Test Score']))
print('\nLasso regression training score : {}'.format(Scores['Lasso Train Score']))
print('Lasso regression testing score  : {}'.format(Scores['Lasso Test Score']))
```

```python
#print regression coefficents for each method
print('Predicting: {}'.format(objective1))
print('Coeffecient names:\n{}'.format(coefs1))
print('Coeffecients for Linear Regression: \n{}'.format(lin_reg.coef_))
print('Coeffecients for Ridge Regression: \n{}'.format(rig_reg.coef_))
print('Coeffecients for Lasso Regression: \n{}'.format(las_reg.coef_))
```

```python
if Scores['Lasso Test Score'] > Scores['Ridge Test Score']:
    best = 'Lasso'
    score = Scores['Lasso Test Score']
else:
    best = 'Ridge'
    score = Scores['Ridge Test Score']

print('Between Ridge and Lasso, the best ML algorithm is {}, with a score of {}.'.format(best,score))
```
