```python
# 7-6 Three Exits:
#conditional test loop
quit = False
toppings = []

while quit == False:
    new = input('Please input a pizza topping. \nIf you are done entering toppings please enter "quit".\n\t')
    
    if str.strip(new.lower()) == 'quit':
        quit = True
    else:
        toppings.append(str.strip(new.lower()))
        print("I'm adding " + str(new) + " to your pizza!\n")
```

```python
#7-9 No Pastrami:
sandwich_orders = ['pb&j','Pastrami','reuben','  pastrami ','grilled cheese','pastrami']
finished_sandwiches = []
message = False

for sandwich in sandwich_orders:
    if str.strip(sandwich.lower()) == 'pastrami':
        if message == False:
            print('Sorry we are out of pastrami.\n')
            message = True
        sandwich_orders.remove(sandwich)
            
for sandwich in sandwich_orders:
    print('I made your ' + sandwich + '.\n')
    finished_sandwiches.append(str.strip(sandwich.title()))
    
print('The following sandwiches have been made:\n' + str(finished_sandwiches))
```

```python
#8-13 User Profile:
def build_profile(first, last, **user_info):
    """Build a dictionary containing everything we know about a user."""
    profile = {}
    profile['first_name'] = first
    profile['last_name'] = last
    for key, value in user_info.items():
        profile[key] = value
    return profile

user_profile = build_profile('james', 'sloan',
                            location='clemson',
                            field='IE',
                            hobby='climbing')
print(user_profile)
```

```python
#8-14 Cars:
def cars(manufacturer, model_name, **additional_details):
    car_profile = {}
    car_profile['manufacturer'] = manufacturer
    car_profile['model name'] = model_name
    for key, value in additional_details.items():
        car_profile[key] = value
    return car_profile

car = cars('subaru','outback',color = 'blue', tow_package = True)
print(car)
```
